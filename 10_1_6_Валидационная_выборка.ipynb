Дата-саентисты подкрутили немного параметры сети и для контроля подали на вход нейросети **такую же** валидационную выборку (ВВ), что и в предыдущей задаче. Оказалось, что теперь уже четверо истинно больных были классифицированы как больные, и precision для класса 1 оказалась равна  0.4

Как видно, precision и recall **возросли по сравнению с прошлой задачей**. Но следует ли из этого, стало меньше здоровых людей, которые классифицируются нейросетью как больные?

# Решение

$ TP = 4 $ (четверо истинно больных были классифицированы как больные)

$ Precision = 0.4 $ (precision для класса 1 оказалась равна  0.4)

Поскольку $precision = TP / (TP + FP)$, подставив $0.4 = 4 / (4 + FP)$

Получаем $ FP = (4 - 1.6) / 0.4 = 6$
