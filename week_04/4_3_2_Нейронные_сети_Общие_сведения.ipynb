Представьте себе такую многослойную нейронную сеть: у нее входной слой размерности 4, далее идут внутренние полносвязные слои, состоящие соответственно из 10, 20, 5 нейронов. Вся сеть заканчивается выходным слоем из одного нейрона.

Сколько в этой сети весов-связей и весов-смещений? Напишите в ответ общее количество весов всех типов.

```python
# 381
```

Входной слой (4 нейрона) -> Слой 1 (10 нейронов) -> Слой 2 (20 нейронов) -> Слой 3 (5 нейронов) -> Выходной слой (1 нейрон)

* **Связи от слоя 0 к слою 1:** 4 нейрона входа * 10 нейронов слоя 1 = 40 весов
* **Связи от слоя 1 к слою 2:** 10 нейронов слоя 1 * 20 нейронов слоя 2 = 200 весов
* **Связи от слоя 2 к слою 3:** 20 нейронов слоя 2 * 5 нейронов слоя 3 = 100 весов
* **Связи от слоя 3 к слою 4:** 5 нейронов слоя 3 * 1 нейрон слоя 4 = 5 весов

**Всего весов-связей:** 40 + 200 + 100 + 5 = 345 весов

* **Смещения для слоя 1:** 10 смещений
* **Смещения для слоя 2:** 20 смещений
* **Смещения для слоя 3:** 5 смещений
* **Смещения для слоя 4:** 1 смещение

**Всего весов-смещений:** 10 + 20 + 5 + 1 = 36 весов

**Общее количество весов всех типов:** 345 + 36 = 381


```python
s = 4, 10, 20, 5, 1
sum(s[i-1]*s[i] + s[i] for i in range(1, len(s)))
```
