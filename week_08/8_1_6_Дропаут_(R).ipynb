```r
#@title Со сдвигом маски на каждой итерации
d <- data.frame(X1 = c(0, 0, 1, 1),
                X2 = c(0, 1, 0, 1),
                Y1 = c(0, 0, 0, 1))
C <- 1
h <- 0.1
w <- c(0, 0)
bias <- 0
mask <- c(1, 0)
epoches <- 200

# Сдвиг маски:
shift_lst <- function(lst, shift) {
    if (shift > 0) {
        c(tail(lst, -shift), head(lst, shift))
    } else {
        lst # Не сдвигаем, если shift нулевой или отриц.
    }
}


# Предсказание модели:
y_pred <- function(x) {
  mask[1] * w[1] * x[1] + mask[2] * w[2] * x[2] + bias
}

# Цикл по эпохам:
for (epoch in 1:epoches) {
    # Подсчёт градиента:
    grad <- c(0, 0)
    grad_bias <- 0
    for (i in 1:nrow(d)) {
        error <- y_pred(unlist(d[i, c("X1", "X2")])) - d$Y1[i]
        grad <- grad + 2 * c(error * d$X1[i] * mask[1], error * d$X2[i] * mask[2])
        grad_bias <- grad_bias + 2 * error
    }
    # Обновление весов w и bias:
    w <- w - h * grad
    bias <- bias - h * grad_bias
    # Результаты эпохи:
    cat('epoch =', epoch, 'mask =', mask, 'w_0 =', round(bias, 2), 'w_1 =', round(w[1], 2), 'w_2 =', round(w[2], 2), '\n')
    # Смена маски
    mask <- shift_lst(mask, shift=1)
}
```
