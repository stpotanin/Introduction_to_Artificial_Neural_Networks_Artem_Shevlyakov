Допустим, что наша тренировочная выборка состоит из 4х объектов. Три объекта А1, А2, А3 в реальности принадлежат классу 0, а один объект А4 в реальности принадлежит классу 1.

| A | class |
|---|---|
| A1 | 0 |
| A2 | 0 |
| A3 | 0 |
| A4 | 1 |

Как будет работать дурацкий классификатор К1 на такой выборке? Подсчитайте его функцию потерь.

А теперь рассмотрим другой (не-дурацкий) классификатор К2. Он для объектов А1, А2, А3 выдал вероятность принадлежности классу 0 равную 0.6 . Чему равна вероятность принадлежности объекта А4 к классу 1 по мнению классификатора К2, если известно, что функция потерь К2 на всей тренировочной выборке численно равна функции потерь классификатора К1?  



**Спойлер**. Если вы всё сделали правильно, то функция потерь выше у тренировочной выборки классификатора К2. Это означает, что К2 "более глуп", чем К1. Дело в том, что К2 имел в своем распоряжении бОльшую выборку, чем К1, но ничего кроме дурацкой стратегии классификации из нее не извлёк. Поэтому и функция потерь у него больше.

```python
from sympy import log, exp, symbols, solve

x = symbols('x')
round(exp(solve(3 * log(.75) + log(.25) - 3 * log(.6) - x)[0]), 2)
```
